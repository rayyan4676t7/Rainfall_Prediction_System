Weather Rainfall Prediction System Using Neural Network Analysis

Problem Statement

Accurate rainfall prediction plays a crucial role in agriculture, water resource management, disaster preparedness, and environmental planning. Traditional forecasting methods may struggle to capture complex relationships between weather parameters. This project aims to develop a machine learning-based rainfall prediction system using a neural network model to estimate rainfall based on historical weather conditions.

Objective

The objective of this project is to predict rainfall using meteorological parameters such as temperature, humidity, wind speed, pressure, cloud cover, month, and year. The project utilizes a feed-forward neural network to learn patterns from historical weather data and generate rainfall predictions.

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

## Dataset Description

The dataset contains historical weather observations with the following features:

* Year
* Month
* Temperature
* Humidity
* Wind Speed
* Pressure
* Cloud Cover
* Rainfall (Target Variable)

```python
df = pd.read_csv("rainfall_dataset.csv")

print(df.head())
print(df.shape)

df.fillna(df.mean(numeric_only=True), inplace=True)

X = df.iloc[:, :-1]
y = df.iloc[:, -1:]
```

### Data Preprocessing

* Loaded the dataset using Pandas.
* Handled missing values using mean imputation.
* Split the dataset into input features and target variables.
* Applied MinMaxScaler for feature normalization.
* Performed an 80-20 train-test split.

```python
scaler_x = MinMaxScaler()
scaler_y = MinMaxScaler()

X_scaled = scaler_x.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y_scaled,
    test_size=0.2,
    random_state=42
)
```

### Neural Network Architecture

The rainfall prediction model consists of:

* Input Layer: 7 Features
* Hidden Layer 1: 32 Neurons (ReLU Activation)
* Hidden Layer 2: 16 Neurons (ReLU Activation)
* Output Layer: 1 Neuron (Rainfall Prediction)

Optimizer:

* Adam

Loss Function:

* Mean Squared Error (MSE)

Training Epochs:

* 60


```python
model = Sequential([
    Dense(32, activation="relu", input_shape=(X.shape[1],)),
    Dense(16, activation="relu"),
    Dense(1)
])

model.compile(
    optimizer="adam",
    loss="mse"
)

history = model.fit(
    X_train,
    y_train,
    epochs=60,
    verbose=1
)
```


## Results

Model Performance:

* RВІ Score: 0.8289
* Prediction Performance: 82.89%
* Mean Absolute Error (MAE): 2.63 mm
* Root Mean Squared Error (RMSE): 3.04 mm

The model demonstrated strong predictive capability by successfully learning relationships between weather parameters and rainfall levels.

```python
preds = model.predict(X_test)

preds_rescaled = scaler_y.inverse_transform(preds)
y_test_rescaled = scaler_y.inverse_transform(y_test)

r2 = r2_score(y_test_rescaled, preds_rescaled)
mae = mean_absolute_error(y_test_rescaled, preds_rescaled)
rmse = np.sqrt(mean_squared_error(y_test_rescaled, preds_rescaled))

print("RВІ Score:", r2)
print("Accuracy (%):", r2 * 100)
print("MAE:", mae)
print("RMSE:", rmse)
```
